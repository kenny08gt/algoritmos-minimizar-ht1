# 1.- Dada la funcióon 
\begin{align}
f: R → R_{3}
\end{align}

\begin{align}
f(x) = x^4 - 6x^3 + 3x^2 + 10x
\end{align}

## a) Utilice cualquier software para graficar dicha función.

**x de -10 a 15 con incrementos de 1**

```python
import matplotlib.pyplot as plt
import numpy as np

def value_function(x):
    y = np.zeros(x.shape[0])
    for i in range(x.shape[0]):
        y[i] = x[i]**4 - 6 * (x[i]**3) + 3*(x[i]**2) + 10*x[i]
    return y

x = np.arange(-150, 150)
# print(x.shape[0])
y = value_function(x)
plt.plot(x, y, linewidth=2.0)
plt.show()
```


**x de -2 a 6, con incrementos de 0.1**

```python
x = np.arange(-2, 6, step = 0.1)
# print(x.shape[0])
y = value_function(x)
plt.plot(x, y, linewidth=2.0)
plt.show()
```

## b) Localice los máximos y mínimos locales, si los hay.


```python
local_x1 = np.arange(-2, 0, step = 0.1)
local_y1 = value_function(local_x1)

local_min1_y = np.min(local_y1)
local_min1_x = local_x1[np.argmin(local_y1)]

local_x2 = np.arange(3, 5, step = 0.1)
local_y2 = value_function(local_x2)

local_min2_y = np.min(local_y2)
local_min2_x = local_x2[np.argmin(local_y2)]


local_x3 = np.arange(0, 2, step = 0.1)
local_y3 = value_function(local_x3)

local_min3_y = np.max(local_y3)
local_min3_x = local_x3[np.argmax(local_y3)]

plt.plot(x, y, linewidth=2.0)
plt.scatter(local_min1_x, local_min1_y, marker='x', color='red')
plt.text(local_min1_x - 0.3, local_min1_y + 8, 'Local min', fontsize=9)

plt.scatter(local_min2_x, local_min2_y, marker='x', color='red')
plt.text(local_min2_x - 0.65, local_min2_y + 8, 'local min', fontsize=9)

plt.scatter(local_min3_x, local_min3_y, marker='x', color='red')
plt.text(local_min3_x - 0.3, local_min3_y + 5, 'local max', fontsize=9)

plt.show()
```

## c) ¿Existe un mínimo global? ¿y un máximo global? justifique su respuesta.

```python
global_min_y = np.min(y)
global_min_x = x[np.argmin(y)]

for i in range(x.shape[0]):
    y[i] = x[i]**4 - 6 * (x[i]**3) + 3*(x[i]**2) + 10*x[i]
plt.plot(x, y, linewidth=2.0)
plt.scatter(global_min_x, global_min_y, marker='x', color='red')
plt.text(global_min_x - 0.65, global_min_y + 8, 'Global min', fontsize=9)
plt.show()
```

## c) Utilizando cualquier software resuelva el problema de optimización del inciso anterior

```python
import scipy.optimize as opt

def f(x):
    return x[0] + x[1] * 2.6086997e-16

x0 = [39, 165]

cons=({'type': 'ineq',
       'fun': lambda x: x[0]},
      {'type': 'ineq',
       'fun': lambda x: x[1]})

res = opt.minimize(f, x0, constraints=cons)
print(res)
```
